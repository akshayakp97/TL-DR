```python
import os
import json
import string
import subprocess
import re
import pandas, numpy, string
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
import pickle
```

```python
violence_wordList = list()
# 'with' can automate finish 'open' and 'close' file
with open("/home/raj/UB_Stuff/CSE_635/Phase5/Violence_IncludeWords_Strict.txt") as f:
    # fetch one line each time, include '\n'
    for line in f:
        # strip '\n', then append it to wordlist
        violence_wordList.append(line.rstrip('\n'))
        
protest_wordList = list()
# 'with' can automate finish 'open' and 'close' file
with open("/home/raj/UB_Stuff/CSE_635/Phase5/Protest_IncludeWords_Strict.txt") as f:
    # fetch one line each time, include '\n'
    for line in f:
        # strip '\n', then append it to wordlist
        protest_wordList.append(line.rstrip('\n'))
        
politics_wordList = list()
# 'with' can automate finish 'open' and 'close' file
with open("/home/raj/UB_Stuff/CSE_635/Phase5/politics_India.txt") as f:
    # fetch one line each time, include '\n'
    for line in f:
        # strip '\n', then append it to wordlist
        politics_wordList.append(line.rstrip('\n'))

politics_violence_wordList = list()
# 'with' can automate finish 'open' and 'close' file
with open("/home/raj/UB_Stuff/CSE_635/Phase5/Violence_IncludeWords_political_violent.txt") as f:
    # fetch one line each time, include '\n'
    for line in f:
        # strip '\n', then append it to wordlist
        politics_violence_wordList.append(line.rstrip('\n'))
```

```python
def isProtestWord(sentence):
    trueCount = 0
    wordList = sentence.split(' ')
    for word in wordList:
        if(word in protest_wordList):
            trueCount = trueCount+1;
    
    if trueCount >= 2:
        return True
    else:
        return False
    
```

```python
def isViolenceWord(sentence):
    trueCount = 0
    flag = False
    wordList = sentence.split(' ')
    for word in wordList:
        if(word in violence_wordList):
            trueCount = trueCount+1;
    
    if trueCount >= 3:
        return True
    else:
        return False
```

```python
def isPolitical(sentence):
    trueCount = 0
    flag = False
    wordList = sentence.split(' ')
    for word in wordList:
        if(word in politics_wordList):
            trueCount = trueCount+1;
    
    if trueCount >= 1:
        return True
    else:
        return False
```

```python
def isPoliticalViolence(sentence):
    trueCount = 0
    flag = False
    wordList = sentence.split(' ')
    for word in wordList:
        if(word in politics_violence_wordList):
            trueCount = trueCount+1;
    
    if trueCount >= 2:
        return True
    else:
        return False
```

```python
def getPoliticalViolence(sentence):
    retlist = []
    wordList = sentence.split(' ')
    for word in wordList:
        if(word in politics_violence_wordList):
            retlist.append(word)
    
    return retlist
```
