```python
def Nmaxelements(list1):
    num = -1.0
    final_list = []
    word_list = []

    for i in range(0, len(list1)):
        for k in list1[i]:
            final_list.append(float(list1[i][k]))
   
    for i in range(0, len(final_list)):  
        num = -1.0
        ind = -1
        for j in range(len(final_list)):
            if final_list[j] > num: 
                num = final_list[j]
                ind = j
        for key in list1[ind]:
            word_list.append(key)
        del list1[ind]
        del final_list[ind]
    return word_list
```
